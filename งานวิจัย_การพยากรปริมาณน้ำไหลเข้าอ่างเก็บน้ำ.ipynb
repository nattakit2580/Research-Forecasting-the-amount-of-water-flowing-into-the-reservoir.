# เฉพาะคอลัมน์  inflow  inflow_7h ไปใส่ไว้ใน hydro_data ต่อจากคอลัมน์เดิมที่มีอยู่

```python
# อ่านข้อมูลจากไฟล์ dam_reindex.csv
dam_data = pd.read_csv("dam_reindex.csv")

# อ่านข้อมูลจากไฟล์ hydro_reindex.csv
hydro_data = pd.read_csv("hydro_reindex.csv")

# รวมข้อมูลโดยใช้คอลัมน์ hourlydate เป็นตัวระบุข้อมูลเชื่อมโยง
hydro_data = pd.merge(hydro_data, dam_data[['hourlydate', 'inflow', 'inflow_7h']], on='hourlydate')

# แสดงผลลัพธ์
print("ข้อมูล hydro_data หลังรวมเพิ่มคอลัมน์ inflow และ inflow_7h:")
hydro_data
```

# Feature Selection เพื่อเลือกคุณลักษณะหรือคุณสมบัติที่มีผลต่อการทำนายของโมเดลอย่างมีประสิทธิภาพมากที่สุด

```python
from sklearn.feature_selection import SelectKBest, f_regression

# เลือก features ที่ต้องการใช้ในการทำ Feature Selection
features = ['E65Q', 'E67Q', 'E88Q', 'E89Q', 'E76AQ', 'E90Q', 'inflow']

# สร้าง DataFrame ของ features
X = hydro_data[features]

# สร้าง Series ของ target variable
y = hydro_data['inflow_7h']

# เลือกคุณลักษณะที่ดีที่สุด โดยใช้ SelectKBest และค่า F-score
selector = SelectKBest(score_func=f_regression, k='all')  # เลือกค่า k ตามต้องการ

# ทำการ Fit และ Transform ข้อมูล
X_new = selector.fit_transform(X, y)

# แสดงค่า F-score และคำนวณค่า p-value ของแต่ละคุณลักษณะ
feature_scores = pd.DataFrame({'Feature': X.columns, 'F-score': selector.scores_, 'p-value': selector.pvalues_})
print(feature_scores)

# พล็อตกราฟแสดงความสำคัญของคุณลักษณะ
plt.figure(figsize=(10, 6))
plt.barh(feature_scores['Feature'], feature_scores['F-score'], color='skyblue')
plt.xlabel('F-score')
plt.ylabel('Feature')
plt.title('Feature Importance based on F-score')
plt.show()
```

จากโค้ดนี้จะได้ตารางแสดงค่า F-score และ p-value ของแต่ละคุณลักษณะ พร้อมทั้งพล็อตกราฟแสดงความสำคัญของคุณลักษณะด้วยค่า F-score ครับ


*   โดยค่า F-score ที่สูงก็แสดงถึงคุณลักษณะที่มีผลต่อการทำนายโมเดลได้ดีมากที่สุด
*  p-value ที่น้อยก็แสดงถึงความสัมพันธ์ที่มีนัยสำคัญระหว่างคุณลักษณะกับตัวแปรเป้าหมาย หาก p-value มีค่าน้อยกว่าค่าเข้าสู่เปรียบเทียบ แสดงว่าคุณลักษณะนั้นมีความสำคัญที่มีนัยสำคัญต่อการทำนายโมเดลครับ หากต้องการเลือกคุณลักษณะที่ดีที่สุดสามารถใช้ค่า F-score ที่สูงที่สุดเป็นเกณฑ์ครับ
จะเห็นได้เลยว่าข้อมูลที่สำคัญเป็นข้อมูลของ ['E65Q', 'E67Q', 'E88Q', 'inflow'] จริง ๆ ด้วย

# เลือก features ที่สำคัญเพื่อไปใช้ในการสร้างโมเดลของเราโดยดูจากการทำ Feature Selection

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# เลือก features
features = ['E65Q', 'E67Q', 'E88Q',	'inflow']

# เตรียมข้อมูล
X = hydro_data[features]
y = hydro_data['inflow_7h']

# แบ่งข้อมูลเป็นชุดฝึกและชุดทดสอบ
train_size = int(0.8 * len(hydro_data))
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]
```



*   เลือก features: กำหนดคุณลักษณะที่เราสนใจในการใช้ในการทำนาย 'inflow_7h' ซึ่งคือ 'E65Q', 'E67Q', 'E88Q', และ 'inflow' โดยเก็บไว้ในตัวแปร features.

*   เตรียมข้อมูล: โค้ดจะดึงข้อมูลของ features และ target variable ('inflow_7h') จาก hydro_data แล้วนำมาเก็บไว้ในตัวแปร X และ y ตามลำดับ.


*   แบ่งข้อมูล: โค้ดทำการแบ่งข้อมูล X และ y เป็นชุดฝึกและชุดทดสอบ โดยใช้ขนาดของชุดฝึกเป็น 80% ของข้อมูลทั้งหมด และชุดทดสอบเป็น 20% ของข้อมูลทั้งหมด ซึ่งจะนำไปใช้ในการสร้างและทดสอบโมเดล

# ใช้ MinMaxScaler จากไลบรารี sklearn.preprocessing เพื่อทำการปรับค่าของข้อมูลให้อยู่ในช่วง 0 ถึง 1 ซึ่งเรียกว่าการ Normalize ข้อมูล

```python
from sklearn.preprocessing import MinMaxScaler

# สร้าง scaler
scaler = MinMaxScaler()

# Normalize ข้อมูลฝึกและทดสอบ
X_train_normalized = scaler.fit_transform(X_train)
X_test_normalized = scaler.transform(X_test)
```



*   สร้าง scaler: โค้ดสร้าง MinMaxScaler object และเก็บไว้ในตัวแปร scaler เพื่อนำไปใช้ในการ Normalize ข้อมูลภายหลัง

*   Normalize ข้อมูลฝึกและทดสอบ: โดยใช้ method fit_transform() ของ scaler เพื่อ Normalize ข้อมูลในชุดฝึก (X_train) และชุดทดสอบ (X_test) โดยข้อมูลที่ Normalize จะถูกปรับให้อยู่ในช่วง 0 ถึง 1

# สร้างโมเดล Gradient Boosting Regressor ด้วยการกำหนดพารามิเตอร์ของโมเดล เพื่อให้มีค่าที่เหมาะสมสำหรับการฝึกสอนโมเดล

```python
from sklearn.ensemble import GradientBoostingRegressor

# สร้างโมเดล Gradient Boosting Regressor โดยกำหนดพารามิเตอร์
gb_model = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=5, random_state=42)

# บันทึกค่าการฝึกโมเดล
history = gb_model.fit(X_train_normalized, y_train)
```

โดยกำหนดพารามิเตอร์ดังนี้:


*   n_estimators=100: จำนวนต้นไม้ใน ensemble ที่จะใช้ในการฝึกสอนโมเดล

*   learning_rate=0.1: อัตราการเรียนรู้ของโมเดล

*   max_depth=5: ความลึกของต้นไม้ใน ensemble ซึ่งความลึกจะเป็นขีดจำกัดของการสร้างโครงสร้างต้นไม้ ซึ่งเป็นการป้องกันการเกิด overfitting

*   random_state=42: ใช้กำหนดค่า random state เพื่อให้การสุ่มในกระบวนการฝึกสอนโมเดลเป็นไปตามลำดับที่กำหนด



ใช้ข้อมูลฝึก (X_train_normalized, y_train) เพื่อฝึกโมเดล Gradient Boosting Regressor โดยการเรียกใช้เมธอด fit() ของโมเดล ซึ่งโมเดลจะถูกฝึกสอนด้วยข้อมูลฝึกและหาค่าที่เหมาะสมสำหรับการทำนายโมเดลจากข้อมูลที่ให้มา

# นำไปทำนายค่า

```python
# ทำนายค่า
y_pred_gb = gb_model.predict(X_test_normalized)
```

ทำนายค่าโดยใช้โมเดล Gradient Boosting Regressor (gb_model) ที่ฝึกสอนแล้วกับชุดข้อมูลทดสอบ (X_test_normalized) โดยใช้เมธอด predict() ซึ่งจะคืนค่าทำนายเป็น y_pred_gb ซึ่งจะเป็น array ของค่าทำนายสำหรับข้อมูลทดสอบที่รู้มาแล้ว โดยค่าทำนายเหล่านี้จะถูกใช้ในการประเมินประสิทธิภาพของโมเดลโดยเปรียบเทียบกับค่าจริงที่เราทราบมาแล้ว

# วัดประสิทธิภาพของโมเดลด้วย MSE, RMSE, R2 หรือ R-Squared

```python
from sklearn.metrics import mean_squared_error, r2_score

# คำนวณค่า MSE
mse_gb = mean_squared_error(y_test, y_pred_gb)
print("Mean Squared Error (MSE) with Gradient Boosting:", mse_gb)

# คำนวณค่า RMSE
rmse_normalized = mean_squared_error(y_test, y_pred_gb, squared=False)
print("Root Mean Squared Error (RMSE) with Normalization:", rmse_normalized)

# คำนวณค่า R^2
r2_normalized = r2_score(y_test, y_pred_gb)
print("R^2 Score with Normalization:", r2_normalized)
```



*   MSE, RMSE มีค่ายิ่งต่ำยิ่งดี ถ้าเท่ากับ 0 แปลว่าโมเดลทำนายค่า y ได้ถูกต้องเป๊ะ 100% ในทางปฏิบัติโอกาสที่จะเทรนโมเดลได้ loss = 0 เป็นไปได้ยากมาก
*   R2 หรือ R-Squared อธิบายง่ายๆ R2 คือ variance ที่โมเดลของเราอธิบายได้เป็นสัดส่วนจาก total variance ทั้งหมดของข้อมูลชุดนั้น (นักสถิติใช้สูตรคำนวณ total variance = explained variance + error) โดย R2 จะมีค่าอยู่ระหว่าง 0-1 ยิ่งเข้าใกล้ 1 แปลว่าโมเดลเราทำนายผลได้ดีมากเลยครับ

ดังนั้น ผลลัพธ์ที่ได้จากการคำนวณค่า MSE, RMSE, และ R^2 จะช่วยให้เราเข้าใจถึงประสิทธิภาพของโมเดล Gradient Boosting Regressor ในการทำนายค่า 'inflow_7h' จากข้อมูล 'E65Q', 'E67Q', 'E88Q', และ 'inflow' ได้และมีความแม่นยำอย่างไร

# ทดสอบสร้างโมเดล RandomForestRegressor เพื่อนำมาเปรียบเทียบกับโมเดล Gradient Boosting Regressor เพื่อดูว่าโมเดลไหนมีประสิทธิภาพมากกว่านี้ในชุดข้อมูลนี้

```python
from sklearn.ensemble import RandomForestRegressor

# สร้างโมเดล Random Forest Regressor โดยกำหนดพารามิเตอร์
rf_model = RandomForestRegressor(n_estimators=100, max_depth=5, random_state=42)

# บันทึกค่าการฝึกโมเดล
history = rf_model.fit(X_train_normalized, y_train)

from sklearn.metrics import r2_score

# ทำนายค่า
y_pred_rf = rf_model.predict(X_test_normalized)

# คำนวณค่า MSE
mse_rf = mean_squared_error(y_test, y_pred_rf)
print("Mean Squared Error (MSE) with Random Forest:", mse_rf)

# คำนวณค่า RMSE
rmse_rf = mean_squared_error(y_test, y_pred_rf, squared=False)
print("Root Mean Squared Error (RMSE) with Random Forest:", rmse_rf)

# คำนวณค่า R^2
r2_rf = r2_score(y_test, y_pred_rf)
print("R^2 Score with Random Forest:", r2_rf)
```

**จากผลลัพธ์ที่ได้:**
**Mean Squared Error (MSE):**

*   Random Forest: MSE = 11.1253

*   Gradient Boosting: MSE = 11.4526

*   อธิบาย: MSE หมายถึงค่าเฉลี่ยของต่าง ๆ ระหว่างค่าทำนายกับค่าจริง โมเดล Random Forest มี MSE ที่ต่ำกว่า Gradient Boosting ซึ่งแสดงให้เห็นว่า Random Forest อาจมีประสิทธิภาพในการทำนายที่ดีกว่า

**Root Mean Squared Error (RMSE):**

```python
TRAIN_TEST_CUTOFF = '2023-12-02 00:00:00'
TRAIN_VALID_RATIO = 0.75
scalerX= StandardScaler()
# sq_length = 150 # ข้อมูลย้อนหลัง

# ! Semi Final
sq_length = 150 # ข้อมูลย้อนหลัง
```

```python
def recall_m(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
    recall = true_positives / (possible_positives + K.epsilon())
    return recall

def precision_m(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
    precision = true_positives / (predicted_positives + K.epsilon())
    return precision

def f1_m(y_true, y_pred):
    precision = precision_m(y_true, y_pred)
    recall = recall_m(y_true, y_pred)
    return 2*((precision*recall)/(precision+recall+K.epsilon()))

def f1macro(y_true, y_pred):
    f_pos = f1_m(y_true, y_pred)
    # negative version of the data and prediction
    f_neg = f1_m(1-y_true, 1-K.clip(y_pred,0,1))
    return (f_pos + f_neg)/2
```

```python
def cnnpred_2d(seq_len=sq_length, n_features=7, n_filters=(8,8,8), droprate=0.1):
    "2D-CNNpred model according to the paper"
    model = Sequential([
        Input(shape=(seq_len, n_features, 1)),
        Conv2D(n_filters[0], kernel_size=(1, n_features), activation="relu"),
        Conv2D(n_filters[1], kernel_size=(3,1), activation="relu"),
        MaxPool2D(pool_size=(2,1)),
        Conv2D(n_filters[2], kernel_size=(3,1), activation="relu"),
        MaxPool2D(pool_size=(2,1)),
        Flatten(),
        Dropout(droprate),
        Dense(1, activation="tanh")
    ])
    return model
```

```python
def datagen(data, seq_len, batch_size, targetcol, kind):
    "As a generator to produce samples for Keras model"

    batch = []
    icount =0
    while True:
        # Pick one dataframe from the pool
        key = random.choice(list(data.keys()))
        df = data[key]
        input_cols = [c for c in df.columns if c != targetcol]
        index = df.index[df.index < TRAIN_TEST_CUTOFF]
        split = int(len(index) * TRAIN_VALID_RATIO)

        assert split > seq_len, "Training data too small for sequence length {}".format(seq_len)
        if kind == 'train':
            index = index[:split]
        elif kind == 'valid':
            index = index[split:]
        else:
            raise NotImplementedError
        # Pick one position, then clip a sequence length
        while True:
            t = random.choice(index)     # pick one time step
            n = (df.index == t).argmax() # find its position in the dataframebatch_size
            if n-seq_len+1 < 0:
                continue # this sample is not enough for one sequence length
            frame = df.iloc[n-seq_len+1:n+1]
            batch.append([frame[input_cols].values, df.loc[t, targetcol]])
            break

        if len(batch) == batch_size:
            X, y = zip(*batch)
            X, y = np.expand_dims(np.array(X), 3), np.array(y)
            yield X, y
            batch = []
```
